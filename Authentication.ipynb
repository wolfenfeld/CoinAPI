```python
import requests
import datetime
import json
```

```python
test_key = '994ADE99-61EB-4B4A-90C2-07DF7AF84238'
headers = {'X-CoinAPI-Key' : test_key}
base_url = 'https://rest.coinapi.io/v1/'
```

```python
def get_exchanges():
    url=base_url+'exchanges'
    response = requests.get(url, headers=headers)
    return json.loads(response.content)
```

```python
def get_history(exchange='BITSTAMP_SPOT_BTC_USD',time_start='2016-01-01T00:00:00', limit='5'):
    url = base_url+'ohlcv/'+exchange+'/history?period_id=2HRS&time_start='+time_start+'&limit='+limit
    response = requests.get(url, headers=headers)
    return json.loads(response.content)
```
